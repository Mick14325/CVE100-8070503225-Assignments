```python
def average_by_key(rows):
    total = 0
    count = 0
    for r in rows:
        if "score" in r:
            total += r["score"]
            count += 1
    if count == 0:
        return 0
    return total / count

print(average_by_key([{"score": 5}, {"score": 7}, {"x": 1}])) # Expected Output: 6.0
print(average_by_key([]))   # Expected Output: 0
```

```python
def even_numbers(nums):
    evens = [] # Initialize the list to store even numbers
    for n in nums: # Corrected from 'num' to 'nums'
        if n % 2 == 0:
            evens.append(n)
    return evens # Return the list of even numbers

print(even_numbers([1,2,3,4,5,6])) # Expected Output: [2, 4, 6]
```

```python
class Artist:
    def __init__(self, name, fee):
        self.name = name
        self.fee = fee # Changed from __fee to fee for public access as implied by expected output

    def add_charges(self, additional_charges):
        self.fee = self.fee + additional_charges # Corrected to use self.fee

a1 = Artist("Taylor Swift", 50)
a2 = Artist("Lisa", 65)
# Apply charges to specific artist instances to match expected output
a1.add_charges(20)
a2.add_charges(20)
print("Taylor Swift fee=", a1.fee) # Expected Output: 70
print("Lisa fee=", a2.fee) # Expected Output: 85
```
